```python
%cd /content/EmotionalConversionStarGAN/
import os
import numpy as np
import torch
import torch.utils.data as data_utils
from tqdm import tqdm
import yaml

import stargan.my_dataset as my_dataset

from speechbrain.pretrained.interfaces import foreign_class
```

```python
# Download the pretrained emotion classifier
device = "cuda" if torch.cuda.is_available() else "cpu"

classifier = foreign_class(source="speechbrain/emotion-recognition-wav2vec2-IEMOCAP",
                            pymodule_file="custom_interface.py",
                            classname="CustomEncoderWav2vec2Classifier",
                            run_opts={"device": device})

# This creates a folder in the current dir called "pretrained_models"
# Take a look in that directory, specifically at custom_interface.py
# custom_interface.py defines what you can do with the classifier,
# such as call classifier.classify_batch, or classifier.encode_batch
```

```python
# Create the dataloader
dataset_config = yaml.safe_load(open("configs/config_step1.yaml", 'r'))
world_dir = os.path.join(dataset_config["data"]["dataset_dir"], "world")

# Run this to get all the WORLD feature files we have (includes ANGRY emotion)
files = my_dataset.get_filenames(world_dir)

# Or this to get only the test set (does not include ANGRY emotion) 
# _, test_files = my_dataset.get_train_test_split(world_dir, dataset_config)

train_test_split = int(.9 * len(files))

train_dataset = my_dataset.MyDataset(dataset_config, files[:train_test_split])
test_dataset = my_dataset.MyDataset(dataset_config, files[train_test_split:])

train_loader = data_utils.DataLoader(train_dataset, batch_size=32,
                                    collate_fn=my_dataset.collate_length_order,
                                    num_workers=0, shuffle=True)
test_loader = data_utils.DataLoader(test_dataset, batch_size=32,
                                    collate_fn=my_dataset.collate_length_order,
                                    num_workers=0, shuffle=False)
```

# Fine-tune the SER
There's a tutorial: https://colab.research.google.com/drive/1LN7R3U3xneDgDRK2gC5MzGkLysCWxuC3?usp=sharing

```python
from speechbrain.lobes.features import Fbank
import torch 
import speechbrain as sb
import inspect

# Define fine-tuning procedure 
class EncDecFineTune(sb.Brain):

    def on_stage_start(self, stage, epoch):
        self.device = device

        # enable grad for all modules we want to fine-tune
        if stage == sb.Stage.TRAIN:
            # Make wav2vec2 trainable
            self.modules.wav2vec2.freeze = False

            # @eric-zhizu: change modules to what's defined in pretrained_models hyperparams.yml
            for module in [self.modules.wav2vec2, self.modules.avg_pool, self.modules.output_mlp]:
                module.to(device)
                for p in module.parameters():
                    p.requires_grad = True
        
    def compute_forward(self, batch, stage):
        """Forward computations from the waveform batches to the output probabilities."""
        # @eric-zhizu: Decompose input
        _, (wavs, wav_lens), labels = batch

        wavs, wav_lens = wavs.to(self.device), wav_lens.to(self.device)

        # @eric-zhizu: Take a look at https://github.com/speechbrain/speechbrain/blob/develop/speechbrain/lobes/models/huggingface_wav2vec.py
        # for source code of the wav2vec2 model

        # wavs: (batch_size, seq_len)
        feats = self.modules.wav2vec2(wavs)
        # feats: (batch_size, compressed_seq_len, 768) 768 is the number of dimensions
        feats = self.modules.avg_pool(feats, lengths=wav_lens)
        # feats: (batch_size, 1, 768) The avg_pool is the last step in the convolution
        x = self.modules.output_mlp(feats)
        # x: (batch_size, 1, 4)
        x = torch.squeeze(x)

        # x: (batch_size, 4)
        return x

    def compute_objectives(self, predictions, batch, stage):
        """Computes the loss NLL given predictions and targets."""

        p_seq = predictions

        _, _, labels = batch
        emo_labels = labels[:, 0].to(device)

        loss = torch.nn.CrossEntropyLoss()(p_seq, emo_labels)

        return loss

    def fit_batch(self, batch):
        """Train the parameters given a single batch in input"""
        predictions = self.compute_forward(batch, sb.Stage.TRAIN)
        loss = self.compute_objectives(predictions, batch, sb.Stage.TRAIN)
        loss.backward()
        if self.check_gradients(loss):
            self.optimizer.step()
        self.optimizer.zero_grad()
        return loss.detach()
```

```python
import yaml

# @eric-zhizu: Found by inspecting the hyperparams.yaml file in pretrained_models
modules = {"wav2vec2": classifier.mods.wav2vec2,
           "avg_pool": classifier.mods.avg_pool,
           "output_mlp": classifier.mods.output_mlp}

hparam_configs = yaml.load(open('pretrained_models/CustomEncoderWav2vec2Classifier--6682143911803214293/hyperparams.yaml', 'r'), Loader=yaml.BaseLoader)
hparams = {"encoder_dim": hparam_configs['encoder_dim'],
           "out_n_neurons": hparam_configs['out_n_neurons'],
           "label_encoder": hparam_configs['label_encoder'],
           "softmax": hparam_configs['softmax']}

brain = EncDecFineTune(modules, hparams=hparams, opt_class=lambda x: torch.optim.Adam(classifier.parameters(), lr=0.000001, weight_decay = 0.000001))
```

```python
from google.colab import drive
current_epoch = 15
end_epoch = 15

for epoch in range(current_epoch, end_epoch+1):
    brain.fit(range(epoch,epoch+1), train_set=train_loader, valid_set=test_loader)
    torch.save(classifier.state_dict(), f'../gdrive/MyDrive/final_project/checkpoints/finetune_ser_epoch_{epoch}.ckpt')
    # Flush to disk
    drive.mount("/content/gdrive", force_remount=True)
```

# Some example code for using the classifier
This code loops through all samples in the data loader and calculates the accuracy of classifier.

```python
classifier.load_state_dict(torch.load('../gdrive/MyDrive/final_project/checkpoints/finetune_ser_epoch_14.ckpt'))
```

```python
classifier = classifier.to(device)
```

```python
def get_accuracy(model):
    model = model.to(device)
    total_accurate = 0
    total = 0
    for i, (_, (wav_forms, wav_lengths), labels) in tqdm(enumerate(test_loader)):
        if i == 100:
            break
        batch_size = labels.shape[0]

        wav_forms, wav_lengths = wav_forms.to(device), wav_lengths.to(device)

        # predicted_labels: (batch_size, )
        _, _, predicted_labels, _ = classifier.classify_batch(wav_forms, wav_lengths)

        # labels[:,0] emotion labels, labels[:,1] speaker labels
        source_labels = labels[:, 0]
        predicted_labels = predicted_labels.detach().cpu()

        n_accurate = (source_labels == predicted_labels).sum()

        total_accurate += n_accurate
        total += batch_size
    print("Accuracy:", total_accurate / total)
    return total_accurate / total
```

```python
get_accuracy(classifier)
```
